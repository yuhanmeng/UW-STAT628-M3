```python
for i in range(len(data)):
    if data[i]['stars']==starlevel[4]:
        star5.append(data[i]['text'])
    elif data[i]['stars']==starlevel[3]:
        star4.append(data[i]['text'])
    elif data[i]['stars']==starlevel[2]:
        star3.append(data[i]['text'])
    elif data[i]['stars']==starlevel[1]:
        star2.append(data[i]['text'])
    else:
        star1.append(data[i]['text'])
```

```python
star5word=[]
star4word=[]
star3word=[]
star2word=[]
star1word=[]
```

```python
for i in range(len(star5)):
    a=star5[i]
    b=a.replace("[","")
    c=b.replace("]","")
    d=c.split(',')
    for i in range(len(d)):
        e=d[i].replace("'","")
        star5word.append(e)
        
for i in range(len(star4)):
    a=star4[i]
    b=a.replace("[","")
    c=b.replace("]","")
    d=c.split(',')
    for i in range(len(d)):
        e=d[i].replace("'","")
        star4word.append(e)

for i in range(len(star3)):
    a=star3[i]
    b=a.replace("[","")
    c=b.replace("]","")
    d=c.split(',')
    for i in range(len(d)):
        e=d[i].replace("'","")
        star3word.append(e)
        
for i in range(len(star2)):
    a=star2[i]
    b=a.replace("[","")
    c=b.replace("]","")
    d=c.split(',')
    for i in range(len(d)):
        e=d[i].replace("'","")
        star2word.append(e)

for i in range(len(star1)):
    a=star1[i]
    b=a.replace("[","")
    c=b.replace("]","")
    d=c.split(',')
    for i in range(len(d)):
        e=d[i].replace("'","")
        star1word.append(e)
```

```python
for i in range(len(star5word)):
    star5word[i]=star5word[i].strip()
    
for i in range(len(star4word)):
    star4word[i]=star4word[i].strip()   
    
for i in range(len(star3word)):
    star3word[i]=star3word[i].strip()
    
for i in range(len(star2word)):
    star2word[i]=star2word[i].strip()
    
for i in range(len(star1word)):
    star1word[i]=star1word[i].strip()
```

```python
from collections import Counter

d5=Counter(star5word)
d4=Counter(star4word)
d3=Counter(star3word)
d2=Counter(star2word)
d1=Counter(star1word)
```

```python
positiveWordsBarGraph <- function(SC){
  contributions <- SC %>%
    unnest_tokens(word, text) %>%
    count(word,sort = TRUE) %>%
    ungroup() %>%
    
    inner_join(get_sentiments("afinn"), by = "word") %>%
    group_by(word) %>%
    summarize(occurences = n(),
              contribution = sum(score))
  
  contributions %>%
    top_n(20, abs(contribution)) %>%
    mutate(word = reorder(word, contribution)) %>%
    head(20) %>%
    ggplot(aes(word, contribution, fill = contribution > 0)) +
    geom_col(show.legend = FALSE) +
    coord_flip() + theme_bw()
}

positiveWordsBarGraph(wholereview %>%
                        filter(business_id == 'Mem13A3C202RzT53npn4NA'))
```
