```python
for i in range(row):
    tempkey=datamatrix[i]
    keys=[*tempkey]
    for j in range(len(keys)):
        if keys[j] in dellist:
            continue
        index=keyset.index(keys[j])
        keyname=keys[j]
        d[i,index]=int(tempkey[keyname])
```

```python
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_components=5,
                                random_state=628,
                                n_jobs=-1,
                                learning_method='online')
d_topics = lda.fit_transform(d)
```

```python
len(keyset)
```

topic words 

```python
n_top_words = 10
feature_names = keyset

for topic_idx, topic in enumerate(lda.components_):
    print("Topic %d:" % (topic_idx + 1))
    print(" ".join([feature_names[i]
                    for i in topic.argsort()\
                        [:-n_top_words - 1:-1]]))
```
