```python
topic1=data[data.loc[:,"topic"] == 1]
topic2=data[data.loc[:,"topic"] == 2]
topic3=data[data.loc[:,"topic"] == 3]
topic4=data[data.loc[:,"topic"] == 4]
topic5=data[data.loc[:,"topic"] == 5]
```

# Try to sum up the number of adj occurs for each business

```python
names=['overrate', 'food', 'service', 'price']
```

## for topic1

```python
def merge_dicts(d1,d2):
    for i,j in d1.items():
                if i in d2.keys():
                    d2[i] += j
                else:
                    d2[i] = j
    return(d2)
```

```python
def getsum(line):
    newdict=dict()
    for i in names:
        olddict=eval(line.loc[i])
        newdict=merge_dicts(newdict,olddict)
    return(newdict)
    
```

```python
#just try one line
topic1new=dict()
for i in range(3):
    new=dict()
    line=topic1.loc[i,:]
    new['names']=line['Unnamed: 0']
    new['stars']=line['stars']
    adjsum=getsum(line)
    new.update(adjsum)
    topic1new[i]=new
    
```

```python
topic1new=dict()
for i in range(topic1.shape[0]):
    new=dict()
    line=topic1.iloc[i,:]
    new['names']=line['Unnamed: 0']
    new['stars']=line['stars']
    adjsum=getsum(line)
    new.update(adjsum)
    topic1new[i]=new
```

```python
topic1_allsumup=pd.DataFrame(topic1new)
topic1_allsumup=topic1_allsumup.T
```

```python
a=topic1_allsumup.columns.values.tolist()[0:37]+topic1_allsumup.columns.values.tolist()[38:58]+topic1_allsumup.columns.values.tolist()[59:]
order=['names','stars']+a
```

```python
topic1_allsumup=topic1_allsumup[order] #put names and stars at first
```
