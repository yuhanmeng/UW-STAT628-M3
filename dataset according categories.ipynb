```python
import json
with open('business.json', encoding='utf-8') as f:
    iter_f = iter(f)
    line = f.readline()
    text = []
    for line in iter_f: 
        d = json.loads(line)
        text.append(d)
    f.close()
```

```python
text[0]
```

```python
l=list()
for i in range(len(text)):
    s=text[i]['categories']
    if type(s)==str:
        words=s.split(",")
        for word in words:
            l.append(word)
    
l
```

```python
d=dict()
for i in l:
    if i not in d.keys():
        d[i]=1
    else:
        d[i]=d[i]+1
d
```
